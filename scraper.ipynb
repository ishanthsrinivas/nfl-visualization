```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import time
```

```python
seasons = [str(year) for year in range(2020, 2025)]
team_abbrs = ['crd', 'atl', 'rav', 'buf', 'car', 'chi', 'cin', 'cle', 'dal', 'den', 'det', 'gnb', 'htx', 'clt', 'jax', 'kan', 'sdg', 'ram',
             'rai', 'min', 'nwe', 'nor', 'nyg', 'nyj', 'phi', 'pit', 'sea', 'sfo','tam', 'oti', 'was']
```

```python
nfl_df = pd.DataFrame()

for season in seasons:
    
    for team in team_abbrs:
        url = 'https://www.pro-football-reference.com/teams/' + team + '/' + season + '/gamelog/'
        print(url)
        
        offensive_df = pd.read_html(url, header = 1, attrs = {'id': 'gamelog' + season})[0]
        defensive_df = pd.read_html(url, header = 1, attrs = {'id': 'gamelog_opp' + season})[0] 
        team_df = pd.concat([offensive_df, defensive_df], axis = 1)
        
        team_df.insert(loc = 0, column = 'Season', value = season)
        team_df.insert(loc = 1, column = 'Team', value = team.upper())

        nfl_df = pd.concat([nfl_df, team_df], ignore_index = True)
        
                
        time.sleep(random.randint(7, 8))
        
```

```python
import nfl_data_py as nfl
import urllib.request
```

```python
logos = nfl.import_team_desc()
logos = logos[['team_abbr', 'team_logo_espn']]
logos
```

```python
logo_paths = []
team_abbr = []

for team in range(len(logos)):
    urllib.request.urlretrieve(logos['team_logo_espn'][team], f"logos/{logos['team_abbr'][team]}.tif")
    logo_paths.append(f"logos/{logos['team_abbr'][team]}.tif")
    team_abbr.append(logos['team_abbr'][team])
```

```python
data = {'TeamAbbr' : team_abbr, 'Logo Path': logo_paths}
logo_df = pd.DataFrame(data)

logo_df = logo_df.drop([16, 26, 32, 29])
logo_df.reset_index()
```

```python
#How much teams passed and ran the ball in 2023
nfl_df_23 = nfl_df.loc[nfl_df['Season'] == '2023']

nfl_23_run = []
nfl_23_pass = []
for team in team_abbrs:
    nfl_23_run.append(sum(nfl_df_23.loc[nfl_df_23['Team'] == team.upper()]['RushYds']) / len(nfl_df_23.loc[nfl_df_23['Team'] == team.upper()]))
    nfl_23_pass.append(sum(nfl_df_23.loc[nfl_df_23['Team'] == team.upper()]['PassYds']) / len(nfl_df_23.loc[nfl_df_23['Team'] == team.upper()]))
```

```python
passrun_df = pd.DataFrame({
    
    "TeamAbbr": nfl_df_23['Team'].unique(),
    "PassYds": nfl_23_pass,
    "RushYds": nfl_23_run
    
})

team_abbreviation_map = {
    'CRD': 'ARI',  # Cardinals
    'ATL': 'ATL',  # Falcons
    'RAV': 'BAL',  # Ravens
    'BUF': 'BUF',  # Bills
    'CAR': 'CAR',  # Panthers
    'CHI': 'CHI',  # Bears
    'CIN': 'CIN',  # Bengals
    'CLE': 'CLE',  # Browns
    'DAL': 'DAL',  # Cowboys
    'DEN': 'DEN',  # Broncos
    'DET': 'DET',  # Lions
    'GNB': 'GB',   # Packers
    'HTX': 'HOU',  # Texans
    'CLT': 'IND',  # Colts
    'JAX': 'JAX',  # Jaguars
    'KAN': 'KC',   # Chiefs
    'SDG': 'LAC',  # Chargers (formerly San Diego)
    'RAM': 'LAR',  # Rams
    'RAI': 'LV',   # Raiders
    'MIN': 'MIN',  # Vikings
    'NWE': 'NE',   # Patriots
    'NOR': 'NO',   # Saints
    'NYG': 'NYG',  # Giants
    'NYJ': 'NYJ',  # Jets
    'PHI': 'PHI',  # Eagles
    'PIT': 'PIT',  # Steelers
    'SEA': 'SEA',  # Seahawks
    'SFO': 'SF',   # 49ers
    'TAM': 'TB',   # Buccaneers
    'OTI': 'TEN',  # Titans
    'WAS': 'WAS'   # Commanders
}

passrun_df['TeamAbbr'] = passrun_df['TeamAbbr'].replace(team_abbreviation_map)
vis_df = pd.merge(passrun_df, logo_df, on='TeamAbbr')
vis_df
```

```python
def getImage(path):
    if 'NYJ' in path:
        return OffsetImage(plt.imread(path, format = 'tif'), zoom= 0.02)
    return OffsetImage(plt.imread(path, format = 'tif'), zoom= 0.1)


plt.rcParams['figure.figsize'] = [20,14]
plt.rcParams['figure.autolayout'] = True

from matplotlib.offsetbox import OffsetImage, AnnotationBbox

paths = vis_df['Logo Path']

x = vis_df['PassYds']
y = vis_df['RushYds']

fig, ax = plt.subplots()

for x0, y0, path in zip(x, y, paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False)
    ax.add_artist(ab)
    
plt.xlim(min(x) - 20, max(x) + 20)
plt.ylim(min(y) - 20, max(y) + 20)

plt.xlabel('Average Passing Yards', fontsize = 30);
plt.ylabel('Average Rushing Yards', fontsize = 30);
plt.title('Average Passing and Rushing Yards During the 2023 NFL Season', fontsize = 50);
```
